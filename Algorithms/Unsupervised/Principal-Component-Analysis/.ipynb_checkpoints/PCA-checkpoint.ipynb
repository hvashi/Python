## Principal Component Analysis

Principal Component Analysis (PCA) a dimension-reduction tool that can be used to reduce a large set of variables to a small set that still contains most of the information in the large set.

Principal component analysis (PCA) is a mathematical procedure that transforms a number of (possibly) correlated variables into a (smaller) number of uncorrelated variables called principal components. The first principal component accounts for as much of the variability in the data as possible, and each succeeding component accounts for as much of the remaining variability as possible. 

```python
#Loading Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
%matplotlib inline
```

```python
#Gathering Data
credit = pd.read_csv(r"/home/seckindinc/Desktop/Github/Python/Data-Sets/Risk.txt",sep=",",index_col=False)
```

```python
#First 5 Rows of Data
credit.head()
```

```python
#Selecting Numeric Fields To Be Used In Modelling
X = credit[['AGE','INCOME','NUMKIDS','NUMCARDS','LOANS']]
```

```python
#Importing z score scaler library
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X))
```

```python
#Loading Min Max Scaling and Creating New Data Frame
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_normalized = pd.DataFrame(scaler.fit_transform(X))
```

```python
#Assining Column Names
X_scaled.columns = ['AGE','INCOME','NUMKIDS','NUMCARDS','LOANS']
X_normalized.columns = ['AGE','INCOME','NUMKIDS','NUMCARDS','LOANS']
```

```python
#Implementing Algorithm with z score 
from sklearn.decomposition import PCA
pca_scaled = PCA(n_components=2)
pca_scaled.fit(X_scaled)
X_scaled_pca = pca_scaled.transform(X_scaled)
```

```python
#Print shape difference before PCA and after PCA
print('Original data shape: \n',X_scaled.shape)
print('Transformed data shape: \n',X_scaled_pca.shape)
```

```python
#Scatter plot of principal components based on Risk column
plt.figure(figsize=(8,6))
plt.scatter(X_scaled_pca[:,0],X_scaled_pca[:,1],c=np.where(credit['RISK'].str.contains('good'), 1,0))
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#Printing each columns values after PCA
print(pca_scaled.components_)
```

```python
#Heat map of original columns with components
sns.heatmap(pd.DataFrame(pca_scaled.components_,columns=['AGE','INCOME','NUMKIDS','NUMCARDS','LOANS']),cmap='plasma')
```

```python
#Implementing Algorithm with z score 
from sklearn.decomposition import PCA
pca_normalized = PCA(n_components=2)
pca_normalized.fit(X_normalized)
X_normalized_pca = pca_normalized.transform(X_normalized)
```

```python
#Print shape difference before PCA and after PCA
print('Original data shape: \n',X_normalized.shape)
print('Transformed data shape: \n',X_normalized_pca.shape)
```

```python
#Scatter plot of principal components based on Risk column
plt.figure(figsize=(8,6))
plt.scatter(X_normalized_pca[:,0],X_normalized_pca[:,1],c=np.where(credit['RISK'].str.contains('good'), 1,0))
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#Heat map of original columns with components
sns.heatmap(pd.DataFrame(pca_normalized.components_,columns=['AGE','INCOME','NUMKIDS','NUMCARDS','LOANS']),cmap='plasma')
```

```python
#Scatter plot of first principal components based on Risk column
plt.figure(figsize=(8,6))
plt.scatter(X_normalized_pca[:,0],X_scaled_pca[:,0],c=np.where(credit['RISK'].str.contains('good'), 1,0))
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#Scatter plot of second principal components based on Risk column
plt.figure(figsize=(8,6))
plt.scatter(X_normalized_pca[:,1],X_scaled_pca[:,1],c=np.where(credit['RISK'].str.contains('good'), 1,0))
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
#Explained variance ratio among two PCA
print(pca_scaled.explained_variance_ratio_)
print(pca_normalized.explained_variance_ratio_)
```
