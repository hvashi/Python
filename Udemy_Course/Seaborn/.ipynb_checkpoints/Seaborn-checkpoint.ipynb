```python
import seaborn as sns
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
flights = sns.load_dataset("flights")
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],bins=15)
```

Kind parameter lets us to choose inner graph type. Default is scatter. Hex is occurence sensitive. Reg is linear regression line.

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

Hue parameter is used for catagorical column coloring
Palette parameter is color changing

```python
sns.pairplot(data=tips,hue='sex',palette='coolwarm')
```

Estimator is used for estimating the numerical value to analyze. Default is mean. We can use std, etc.

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.mean)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.median)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

Hue parameter lets us to create each graph for each category

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

Violin plot is mixture of boxplot and density plot
Hue parameter lets us to create each graph for each category

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue = 'sex',
              split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True, hue= 'sex')
```

Heat maps are used for color based matrix formation

```python
sns.heatmap(tips.corr(),annot=True,cmap='coolwarm')
```

Flights data set

```python
flights.head()
```

Converting flights data set to matrix form

```python
fp = flights.pivot_table(index='month', columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidths=1)
```

Lmplot is used for linear regression line plotting
Hue is used for category
Markes is used for category labeling via charachters
Scatter_kws is used for label size
Col is used for grid seperator
Row for filtering grids

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,size=8)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
import matplotlib.pyplot as plt
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.violinplot)
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```
