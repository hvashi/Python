```python
#Loading library
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#Gathering Data
credit = pd.read_csv(r"/home/seckindinc/Desktop/Github/Python/Data-Sets/Risk.txt",sep=",",index_col=False)
```

```python
credit.head()
```

```python
#Density Plot
sns.kdeplot(credit['AGE'], shade=True)
```

```python
#Violin Plot
sns.violinplot( y=credit["AGE"] )
```

```python
#Boxplot
sns.boxplot(y=credit["AGE"], x = credit['RISK'], hue = credit['MORTGAGE'] )
```

```python
#Heat Map
# plot it
sns.heatmap(credit[['AGE','NUMCARDS','INCOME','NUMKIDS']].corr(), cmap='PuOr')
```

```python
#Pair Plot
sns.pairplot(credit)
```
